```python
import pandas as pd
```

```python
df_1 = pd.read_csv('data_limpia/datos_limpios_F1.csv')
df_2 = pd.read_csv('data_limpia/datos_limpios_F2.csv')
df_3 = pd.read_csv('data_limpia/datos_limpios_F3.csv')
df_4 = pd.read_csv('data_limpia/datos_limpios_F4.csv')
```

```python
df_1.rename(columns={'marca': 'MARCA', 'modelo': 'MODELO', 'cilindrada': 'CILINDRADA', 'año': 'AÑO'}, inplace=True)
```

```python
df_4.rename(columns={'modelo_base_corregido': 'MODELO', 'version_corregida': 'VERSION', 'cilindrada': 'CILINDRADA', 'AÑO_FAB': 'AÑO'}, inplace=True)
```

```python
# TABLA DE AUTOS

for df in [df_1, df_2, df_3]:
    if "VERSION" not in df.columns:
        df["VERSION"] = None

df_autos = pd.concat([df_1, df_2, df_3, df_4], ignore_index=True)
df_autos.head()
```

```python
# TABLA DE MARCAS

df_marca = pd.DataFrame({
    "ID_MARCA": range(1, len(df_autos["MARCA"].unique()) + 1),
    "MARCA": df_autos["MARCA"].unique()
})

df_marca.head()
```

```python
# Filtrar modelos de una palabra y unir con marca
df_autos["MODELO"] = df_autos["MODELO"].str.upper().str.split().str[0]  # Tomar primera palabra

# Tabla de modelos únicos
df_modelo = df_autos[["MARCA", "MODELO"]].drop_duplicates()
df_modelo = df_modelo.merge(df_marca, left_on="MARCA", right_on="MARCA")
df_modelo = df_modelo[["ID_MARCA", "MODELO"]].rename(columns={"MODELO": "MODELO"})
df_modelo["ID_MODELO"] = range(1, len(df_modelo) + 1)
df_modelo = df_modelo.reindex(['ID_MODELO', 'ID_MARCA', 'MODELO'], axis=1)
```

```python
df_modelo
```

```python
# Tabla modelo (con nombre_marca directo)
df_modelo2 = df_autos[["MARCA", "MODELO"]].drop_duplicates()
df_modelo2["ID_MODELO"] = range(1, len(df_modelo) + 1)
df_modelo2 = df_modelo2.reindex(['ID_MODELO',"MARCA", 'MODELO'], axis=1).reset_index(drop=True)
df_modelo2
```

```python
df_modelo.to_csv('tables/modelos.csv', index=False)
df_autos.to_csv('tables/autos.csv', index=False)
df_marca.to_csv('tables/marcas.csv', index=False)
```
